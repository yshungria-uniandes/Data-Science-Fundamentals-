```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
df.info()
df.head()
```

```python
print('Datos Faltantes por columna:')

print(df.isnull().sum())
```

```python
# Toma el ultimo dato y va a llenar todos los datos Na con la ultima age que se encuentra en el Dtaframe.
df_bfill = df.copy()
df_bfill['age'] = df_bfill['age'].fillna(method='bfill')
# Este codigo te permite
df_bfill.isnull().sum()
```

```python
# Por ejemplo, podemos llenar los valores faltantes en 'age' con la mediana de 'age' por pclass'
# Crear una copia del DataFrame original para no modificarlo directamente
df_mean = df.copy()

# Rellenar valores faltantes (NaN) en la columna 'age'
# 1. Agrupa el DataFrame por la columna 'pclass' (por ejemplo, clase de pasajeros).
# 2. Calcula la media de la columna 'age' para cada grupo ('pclass').
# 3. Usa el método 'transform' para aplicar esa media a cada fila del grupo.
# 4. Reemplaza los valores NaN en 'age' con la media calculada para su respectivo grupo de 'pclass'.
df_mean['age'] = df_mean['age'].fillna(
    df_mean.groupby('pclass')['age'].transform('mean')
)

# Verificar si aún quedan valores faltantes en el DataFrame.
# isnull() identifica valores NaN y sum() los cuenta en cada columna.
df_mean.isnull().sum()
```

```python
# # Por ejemplo, podemos llenar los valores faltantes en 'age' con la mediana de 'age' por pclass'
df_fill_median = df.copy()

df_fill_median['age'] = df_fill_median.groupby('pclass')['age'].transform( lambda x: x.fillna(x.median()))

print(df_fill_median.isnull().sum())
#
```

```python
# Llena los datos nulos que existen en la columna 'edad' con la mediana de la misma columna edad.
df_fill_mean = df.copy()

df_fill_mean['age'] = df_fill_mean['age'].fillna(df_fill_mean['age'].mean())

print("Valores faltantes despues de rellenar 'age' con la media:")

print(df_fill_mean.isnull().sum())
```

```python
# Vamos a borrar todos los registros nulos de la columna 'deck'.
df_drop_na_deck = df.copy()

df_drop_na_deck.dropna(subset=['deck'], inplace=True)

print("Valores faltantes despues de borrar 'deck' con valores nulos:")

print(df_drop_na_deck.isnull().sum())
```

```python
df_drop = df.copy()

df_drop = df_drop.drop(columns = ['deck'])

print("Valores faltantes despues de borrar la columna 'deck' con valores nulos:")

print(df_drop.isnull().sum())
```
